# Displaying results of Karate Social Network

```python
import pandas as pd
import networkx as nx
import matplotlib as mpl
from matplotlib import pyplot as plt
from pylab import rcParams
import networkx as nx
import numpy as np
import os
import sys
import itertools
import re
```

## Experiment with 3 classes and half of the nodes are labeled

```python
with open('karate_labels/part-00000') as f:
    lines = f.readlines()
```

```python
lines = [re.findall(r"(\d.*?),(\d.*?)", i)[0] for i in lines]
l1, l2 = zip(*lines)
l1, l2 = list(l1) ,list(l2)
# Those are node labels
labels = dict(zip(l1,l2))
```

```python
# Those are the possible labels for the problem
possible_labels = set(labels.values())
```

```python
colors = [0,"red", "orange", "green"]
colors_for_nodes = [colors[int(i)] for i in labels.values()]
```

```python
with open("soc-karate/soc-karate.mtx") as f:
    data = f.readlines()
```

```python
# Those are edges between nodes
uselful_data = [re.sub("\n", "",i).split(" ") for i in data[24:]]
uselful_data = [(i[0], i[1]) for i in uselful_data]
```

```python
G = nx.Graph()

for node in labels.keys():
    G.add_node(node)

for edge in uselful_data:
    G.add_edge(*edge)

nx.draw(G, with_labels=True, node_color=colors_for_nodes)
```
